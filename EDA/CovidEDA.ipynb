```python
df2 = newData.copy().set_index('date')
index = ['continent',
         'total_cases_per_million', 
         'new_cases_per_million',
         'total_deaths_per_million']

covid_data_continent = df2[index].reset_index().groupby(['date', 'continent'], as_index=False).mean().set_index('date')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(28, 7))
plt.subplots_adjust(hspace=0.2)

ax1.set(xlabel=" ", ylabel="Total cases per million")
ax1.set_title("Total COVID-19 cases per million by continent", fontsize=15, pad=15)
sns.lineplot(x='date', y='total_cases_per_million',hue='continent', data=covid_data_continent.reset_index(), ax=ax1)

ax2.set(xlabel=" ", ylabel="Total deaths per million")
ax2.set_title("Total deaths per million by continent", fontsize=15, pad=15)
sns.lineplot(x='date', y='total_deaths_per_million',hue='continent', data=covid_data_continent.reset_index(), ax=ax2)

fig.show()
```

```python
latest = newData[newData['date'] == newData["date"].max()]
```

```python
latest_map = latest.groupby(['location'])[["continent", "total_deaths","total_cases","population","extreme_poverty","cardiovasc_death_rate","diabetes_prevalence","life_expectancy","human_development_index"]].sum().reset_index()
```

```python
fig = px.choropleth(latest_map, locations = 'location',
                   locationmode = 'country names',
                   color = "life_expectancy",
                   range_color = [50,90])
fig.update_layout(title = "Life Expectancy Across The World")
fig.show()
```
