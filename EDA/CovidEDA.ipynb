```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None)
data = pd.read_csv('owid-covid-data.csv')
data.head()
```

```python
data.describe()
```

```python
for col in data.columns:
    if data[col].isna().sum() > 300000:
        data.drop(col, inplace=True, axis=1)
data.info()
```

```python
data.hist(figsize=(30, 30), bins=70)
plt.show()
```

```python
df2 = data.copy().set_index('date')
index = ['continent',
         'total_cases_per_million', 
         'new_cases_per_million',
         'total_deaths_per_million']

covid_data_continent = df2[index].reset_index().groupby(['date', 'continent'], as_index=False).mean().set_index('date')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(28, 7))
plt.subplots_adjust(hspace=0.2)

ax1.set(xlabel=" ", ylabel="Total cases per million")
ax1.set_title("Total COVID-19 cases per million by continent", fontsize=15, pad=15)
sns.lineplot(x='date', y='total_cases_per_million',hue='continent', data=covid_data_continent.reset_index(), ax=ax1)

ax2.set(xlabel=" ", ylabel="Total deaths per million")
ax2.set_title("Total deaths per million by continent", fontsize=15, pad=15)
sns.lineplot(x='date', y='total_deaths_per_million',hue='continent', data=covid_data_continent.reset_index(), ax=ax2)

fig.show()
```

# USA data Analysis

```python
data['year'] = pd.DatetimeIndex(data['date']).year
USData = data[data['location'] == 'United States']
USData['date'] = pd.to_datetime(USData['date'])
USData = USData.set_index('date', drop=True)
```

```python
plt.figure(figsize=(15,7))
sns.set(style='darkgrid',)

ax = sns.lineplot(data=USData['new_cases_smoothed']['2020-02-29':])

ax.set(xlabel=" ", ylabel="Number of new cases")
ax.set_title("Daily confirmed COVID-19 cases in USA, 2020-2023", fontsize=15, pad=15)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.set(style='darkgrid',)

ax = sns.lineplot(data=USData['new_deaths_smoothed']['2020-02-29':])

ax.set(xlabel=" ", ylabel="Number of new deaths")
ax.set_title("Daily confirmed COVID-19 deaths in USA, 2020-2023", fontsize=15, pad=15)
plt.show()
```
